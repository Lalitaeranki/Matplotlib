```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')


```

```python
# Read the Mouse and Drug Data
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
mouse_drug_df=pd.read_csv(mouse_drug_data_to_load)
mouse_drug_df.head()
```

```python

#Read  the Clinical Trial Data
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
clinical_trial_df=pd.read_csv(clinical_trial_data_to_load)
clinical_trial_df.columns
```

```python
# Combine the data into a single dataset
combined_df=clinical_trial_df.merge(mouse_drug_df,on="Mouse ID")
combined_df.groupby("Mouse ID").head(1)

# Display the data table for preview
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Mean_tumor_volume=combined_df.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]
Mean_tumor_volume

# 

#Convert to DataFrame 
Mean_tumor_volume_data=Mean_tumor_volume.add_suffix("").reset_index()
#Preview DataFrame
Mean_tumor_volume_data
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error_volume=combined_df.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]
# Convert to DataFrame

# Preview DataFrame
standard_error_volume_data=standard_error_volume.add_suffix("").reset_index()
#Preview DataFrame
standard_error_volume_data
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
Mean_tumor_df=combined_df.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"].unstack(level=0)
Mean_tumor_df
```

```python
# Generate the Plot (with Error Bars)
# Plot sample means with error bars

# Save the Figure
standard_error_volume_df=combined_df.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"].unstack(level=0)

x=standard_error_volume_df.index

fig, ax = plt.subplots()

ax.errorbar(x, Mean_tumor_volume["Capomulin"], standard_error_volume_df["Capomulin"], fmt="--o", color="darkred",
            alpha=0.5, label="Capomulin")
ax.errorbar(x, Mean_tumor_volume["Infubinol"], standard_error_volume_df["Infubinol"], fmt="--^", color="darkblue",
            alpha=0.5, label="Infubinol")
ax.errorbar(x, Mean_tumor_volume["Ketapril"], standard_error_volume_df["Ketapril"], fmt="--s", color="darkgreen",
            alpha=0.5, label="Ketapril")
ax.errorbar(x, Mean_tumor_volume["Placebo"], standard_error_volume_df["Placebo"], fmt="--d", color="black",
            alpha=0.5, label="Placebo")
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_title("Tumor Response to Treatment")
ax.legend(loc="upper left", fontsize="small", fancybox=True)
ax.grid()
plt.tight_layout()
plt.savefig("Tumor response to treatment.png")
plt.show()
```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rate=combined_df.groupby(["Drug","Timepoint"]).count()["Mouse ID"]
survival_rate_df=survival_rate.reset_index()
new_survival_rate_df=survival_rate_df.rename(columns={'Mouse ID':'Mouse Count'})
new_survival_rate_df.head()
# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_rate_df=combined_df.groupby(["Drug","Timepoint"]).count()["Mouse ID"].unstack(level=0)
survival_rate_df
# Preview the Data Frame
```

```python
#survival_rate_df=combined_df.groupby(["Drug","Timepoint"]).count(normalize=True)["Mouse ID"].unstack(level=0)
maximum_survival=max(new_survival_rate_df["Mouse Count"])
survival_rate=(new_survival_rate_df["Mouse Count"]/maximum_survival)*100
survival_rate
per_survival_rate_df=new_survival_rate_df.drop(columns='Mouse Count')
per_survival_rate_df["Survival Rate"]=survival_rate
per_survival_rate_df
df_per_survival = per_survival_rate_df.set_index(['Timepoint','Drug']).unstack()['Survival Rate']
df_per_survival
```

```python
# Generate the Plot (Accounting for percentages)
df_per_survival_final_drugs=pd.DataFrame({"Capomulin":df_per_survival["Capomulin"],"Infubinol":df_per_survival["Infubinol"],
                                          "Ketapril":df_per_survival["Ketapril"],"Placebo":df_per_survival["Placebo"]})
x=df_per_survival.index

fig, ax = plt.subplots()

ax.errorbar(x, df_per_survival_final_drugs["Capomulin"], fmt="--o", color="darkred",
            alpha=0.5, label="Capomulin")
ax.errorbar(x,  df_per_survival_final_drugs["Infubinol"], fmt="--^", color="darkblue",
            alpha=0.5, label="Infubinol")
ax.errorbar(x,  df_per_survival_final_drugs["Ketapril"], fmt="--s", color="darkgreen",
            alpha=0.5, label="Ketapril")
ax.errorbar(x,  df_per_survival_final_drugs["Placebo"], fmt="--d", color="black",
            alpha=0.5, label="Placebo")
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Survival Rate(%)")
ax.set_ylim(min(survival_rate),max(survival_rate)+5)
ax.set_title("Survival During Treatment")
ax.legend(loc="lower left", fontsize="small", fancybox=True)
ax.grid()
plt.tight_layout()
# Save the Figure
plt.savefig("Survival During Treatment.png")



# Show the Figure
plt.show()
df_per_survival_final_drugs
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
cols = ['Capomulin','Ceftamin','Infubinol','Ketapril','Naftisol','Placebo','Propriva','Ramicane','Stelasyn','Zoniferol']
percentage_tumor_change=100*(Mean_tumor_df[cols].iloc[-1]/Mean_tumor_df[cols].iloc[0]-1)
# Display the data to confirm
per_tumor_change=round(percentage_tumor_change)
per_tumor_change
```

```python
# Store all Relevant Percent Changes into a Tuple
df_per_tumor_change=pd.DataFrame({"Drug":cols,"Percentage Change":per_tumor_change})
df_per_tumor_change.reset_index(drop = True, inplace = True)
df_new=df_per_tumor_change.drop(df_per_tumor_change.index[[1,4,6,7,8,9]])
tuples_drug = list(map(tuple,df_new.values))
tuples_drug

# Splice the data between passing and failing drugs

bar_plot=plt.bar(range(len(tuples_drug)), [val[1] for val in tuples_drug], 1,color=['darkred' if val[1] > 0 else "green" for val in tuples_drug], alpha=0.5, align="edge")
    
# Orient widths. Add labels, tick marks, etc

plt.xticks(np.arange(len(tuples_drug))+1,[val[0] for val in tuples_drug])

plt.title("Tumor Change Over 45 Days Treatment")

plt.ylabel("%Tumor Volume Change")

plt.grid()

# Use functions to label the percentages of changes
def autolabel(rects):
    
    for rect in rects:
         height = rect.get_height()
         plt.text(rect.get_x()+rect.get_width()/2., 0.2*height, '%d'%int(height)+'%',ha='center', va='bottom')

# Call functions to implement the function calls
autolabel(bar_plot)

# Save the Figure

plt.savefig("Tumor Change Over 45 days treatment.png")
# Show the Figure
fig.show()
```
