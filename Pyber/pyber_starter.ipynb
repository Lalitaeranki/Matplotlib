```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
# Read the City  Data
city_df=pd.read_csv(city_data_to_load)
#Drop Duplicate entries
city_df=city_df.drop_duplicates('city')
city_df.head()
```

```python
# Read the  Ride Data
ride_data_to_load = "data/ride_data.csv"
ride_df=pd.read_csv(ride_data_to_load)
ride_df.head()
```

```python
# Combine the data into a single dataset
combined_data=ride_df.merge(city_df,on='city')
combined_data

# Display the data table for preview

combined_data.groupby('city').head(1)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#City type gives us list of city types in the dataset(Urban, Suburban, Rural)
city_type=combined_data["type"].unique()
#Calculating the data for Urban type cities for bubble plot average fare per city, total rides per city and drivers per city
urban_data=combined_data.loc[(combined_data["type"]=="Urban")]
Av_fare_urban=urban_data.groupby(["city"])["fare"].mean()
total_rides_urban=urban_data.groupby(["city"])["ride_id"].count()
driver_count_urban_total=urban_data.groupby(["city"])["driver_count"].sum()
driver_count_urban=driver_count_urban_total/total_rides_urban


#Calculating the data for Suburban type cities for bubble plot average fare per city, total rides per city and drivers per city
suburban_data=combined_data.loc[(combined_data["type"]=="Suburban")]
Av_fare_suburban=suburban_data.groupby(["city"])["fare"].mean()
total_rides_suburban=suburban_data.groupby(["city"])["ride_id"].count()
driver_count_suburban_total=suburban_data.groupby(["city"])["driver_count"].sum()
driver_count_suburban=driver_count_suburban_total/total_rides_suburban



#Calculating the data for Rural type cities for bubble plot average fare per city, total rides per city and drivers per city
Rural_data=combined_data.loc[(combined_data["type"]=="Rural")]
Av_fare_Rural=Rural_data.groupby(["city"])["fare"].mean()
total_rides_Rural=Rural_data.groupby(["city"])["ride_id"].count()
driver_count_Rural_total=Rural_data.groupby(["city"])["driver_count"].sum()
driver_count_Rural=driver_count_Rural_total/total_rides_Rural
```

```python
# Show plot
#Scatter plot  and Dataframe for Urban Data
urban_df=pd.DataFrame({"Average Fare Urban":Av_fare_urban,"Total Rides":total_rides_urban,"Driver count":driver_count_urban})

urban_plt=plt.scatter(total_rides_urban, Av_fare_urban,s=driver_count_urban*10,
            marker ='o', facecolors ="orangered", edgecolors='black',alpha = 0.5, label="Urban")

#Scatter plot  and Dataframe for Suburban Data
Suburban_df=pd.DataFrame({"Average Fare Suburban":Av_fare_suburban,"Total Rides":total_rides_suburban,"Driver count":driver_count_suburban})

Suburban_plt=plt.scatter(total_rides_suburban, Av_fare_suburban,s=driver_count_suburban*10,
            marker ='o', facecolors ="lightblue", edgecolors='black',alpha = 0.5, label="Suburban")

#Scatter plot  and Dataframe for rural Data
Rural_df=pd.DataFrame({"Average Fare Rural":Av_fare_Rural,"Total Rides":total_rides_Rural,"Driver count":driver_count_Rural})

Rural_plt=plt.scatter(total_rides_Rural, Av_fare_Rural,s=driver_count_Rural*20,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.5, label="Rural")
#title ,label,legend and te
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number Of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.legend(loc="best",title="City Types",markerscale=0.5)
plt.figtext(1, 0.6, "Note: \nCircle size correlates with driver count per city", fontsize=10)

plt.grid()
plt.tight_layout()
plt.savefig("../Images/Pyber Ride Sharing Data (2016).png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

fare_city_type=combined_data.groupby(["type"]).fare.sum()

labels = ["Rural", "Suburban", "Urban"]

explode = (0, 0, 0.1)

colors=["gold","lightblue","lightcoral"]

# Build Pie Chart
plt.pie(fare_city_type, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")
plt.title("% Total Fare by City Type")

# Save Figure

plt.savefig("../Images/%Total Fare By City Type.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

ride_city_type=combined_data.groupby(["type"]).ride_id.count()

labels = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)
colors=["gold","lightblue","lightcoral"]


# Build Pie Chart
plt.pie(ride_city_type, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")
plt.title("% Total Rides by City Type")

# Save Figure

plt.savefig("../Images/%Total Rides By City Type.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents


driver_city_type=city_df.groupby(["type"]).driver_count.sum()

labels = ["Rural", "Suburban", "Urban"]

explode = (0, 0, 0.1)

colors=["gold","lightblue","lightcoral"]

# Build Pie Chart

plt.pie(driver_city_type, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")
plt.title("% Total Drivers by City Type")

# Save Figure

plt.savefig("../Images/%Total Drivers By City Type.png")

# Show Figure
plt.show()
```
